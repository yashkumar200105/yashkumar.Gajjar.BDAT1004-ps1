```python
#question 3
 "Calculate the area of a triangle using Heron's formula."
  [
   "cell_type": "code",
   "execution_count": null,
   "id": "66529e0b",
   "metadata": {},
   "outputs": [],
   "source": [
    "\n",
    "# Heron's formula for area of a triangle\n",
    "def triangleArea(a, b, c):\n",
    "    s = (a + b + c) / 2\n",
    "    return (s * (s - a) * (s - b) * (s - c)) ** 0.5\n",
    "\n",
    "# Example:\n",
    "print(triangleArea(2, 2, 2))  # 1.7320508075688772\n"
   ]
```

```python
#question 4
 "Separate odd and even numbers."
  [
   "cell_type": "code",
   "execution_count": null,
   "id": "8dd11a3d",
   "metadata": {},
   "outputs": [],
   "source": [
    "\n",
    "# Separate odd and even numbers\n",
    "def separate_odd_even(numbers):\n",
    "    odd = []\n",
    "    even = []\n",
    "    for num in numbers:\n",
    "        if num % 2 == 0:\n",
    "            even.append(num)\n",
    "        else:\n",
    "            odd.append(num)\n",
    "    return odd, even\n",
    "\n",
    "# Example:\n",
    "numbers = [25, 47, 42, 56, 32]\n",
    "odd, even = separate_odd_even(numbers)\n",
    "print(\"Odd numbers:\", odd)   # [25, 47]\n",
    "print(\"Even numbers:\", even) # [42, 56, 32]\n"
   ]
```

```python
#question 10
 "Count letter frequencies in a string."
  [
   "cell_type": "code",
   "execution_count": null,
   "id": "d8297bec",
   "metadata": {},
   "outputs": [],
   "source": [
    "\n",
    "# Count letter frequencies in a string\n",
    "def frequencies(text):\n",
    "    letters = 'abcdefghijklmnopqrstuvwxyz'\n",
    "    freq = [0] * 26\n",
    "    for char in text.lower():\n",
    "        if char in letters:\n",
    "            index = letters.index(char)\n",
    "            freq[index] += 1\n",
    "    return freq\n",
    "\n",
    "# Example:\n",
    "print(frequencies('The quick red fox got bored and went home.'))\n",
    "print(frequencies('apple'))\n"
   ]
```
